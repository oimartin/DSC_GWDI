```python
# Import dependencies
import pandas as pd

# Save csv file name to 'file'
file = "GWDI_orders_coding.csv"

# Import csv file as df
gwdi_orders = pd.read_csv(file)

# Preview df
# gwdi_orders.head()
```

```python
# Save csv file name to 'file'
file_2 = "GWDI_all_mutants.csv"

# Import csv file as df
gwdi_info = pd.read_csv(file_2)

# Preview df
# gwdi_info.head()
```

```python
# Rename strains ordered column of 
gwdi_orders.rename(columns={"Strains Ordered": "ID"}, inplace=True)
# gwdi_orders.head()
```

```python
# Merge GWDI info and orders dfs
gwdi_info_and_orders = gwdi_orders.merge(gwdi_info.set_index('ID'), on='ID', how="inner")
# gwdi_info_and_orders.head()
```

```python
list(gwdi_info_and_orders.columns.values)
```

```python
# Drop unecessary columns
cleaned_gwdi = gwdi_info_and_orders.drop(["Plated", "Form Spores?", "Replated", "Shipped", 
                                          "Harvest Plate","Stored", "Tested", "Unnamed: 14", 
                                          "Chromosome", "Position",], axis=1)
cleaned_gwdi.head()
```

```python
# Looking at different values in 'Disruption' column
cleaned_gwdi['Disruption'].value_counts()
```

```python
# Set index to 'Disruption' and then choose only rows with 'intragenic'
intragenic_only = cleaned_gwdi.set_index('Disruption').loc['intragenic']
```

```python
# Consildated list of wells with intragenic mutant strains
intragenic_only.groupby(['ID']).mean()
```
